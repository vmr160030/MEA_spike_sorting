```python
import numpy as np
import matplotlib.pyplot as plt
import raw
```

## 512 multi-electrode array geometry

```python
# Load electrode map
e_map = np.loadtxt('512_electrode_coords.txt')

# Plot electrode map
plt.scatter(*e_map.T)
plt.xlabel('Electrode x-coordinate (µm)')
plt.ylabel('Electrode y-coordinate (µm)')
```

## Load example snippet of raw data

```python
binpath = '/Volumes/Vyom MEA/data/raw/20250306C/data026'
rt = raw.RawTraces(binpath)

# Load first 20k samples ie 1s
rt.load_bin_data(0, 20000, verbose=True)
```

```python
# Plot channel 0 data
plt.figure(figsize=(12, 4))
time = np.arange(rt.data.shape[1]) / rt.sample_rate
plt.plot(time, rt.data[0])
plt.title('Raw Data from Channel 0')
plt.xlabel('Time (s)')
```
